# Pandas

```python
import pandas as pd
import pandas_datareader as pdr
import datetime
import matplotlib.pyplot as plt
```

```python
appl = pdr.get_data_yahoo('AAPL', start = datetime.datetime(2008,1,1), end = datetime.datetime(2012,1,1))
```

```python
appl.describe()
```

```python
appl['Change'] = appl.Open - appl.Close
```

```python
appl['Pct_Change'] =  appl.Change / appl.Open
```

```python
plt.style.use(['dark_background'])
#plt.patch.set_facecolor('lightgrey')
appl['Pct_Change'].plot(grid=True)
#
```

# Multiple Stocks

```python
def get(tikers, startdate, enddate):
    def data(ticker):
        return (pdr.get_data_yahoo(ticker, start=startdate, end=enddate))
    dataTikers = map(data, tickers)
    return (pd.concat(dataTikers, keys=tickers, names=['Ticker','Date']))
```

```python
tickers = ['MSFT','CRM','GE','MMM']
dataDF = get(tickers, datetime.datetime(2015,1,1),datetime.datetime(2016,1,1))
```

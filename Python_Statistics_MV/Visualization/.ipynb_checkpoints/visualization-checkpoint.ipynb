```python
def plotyear(year):
    data = gapminder[gapminder.Year == year]
    area = 5e-6 * data.population
    colors = data.region.map({'America':'skyblue', 
                              'East Asia & Pacific':'gold', 
                              'Europe & Central Asia':'palegreen',
                              'Middle East & North Africa':'coral', 
                              'South Asia':'red', 'Sub-Saharan Africa':'blue'})
    data.plot.scatter('fertility','child_mortality',s=area,c=colors,
                     linewidths=1,edgecolors='k',figsize=(12,9))
    plt.xlabel('fertility')
    plt.ylabel('mortality')
```

```python
plotyear(1965)
```

```python
interact(plotyear,year=widgets.IntSlider(min=1950,max=2015,step=1,values=1965))
```

```python
size=np.where(italy.Year % 10 == 0,30,2)
italy.plot.scatter("gdp","life",logy=True,s=size)
```

```python
data = gapminder.query('(Country=="Italy") or (Country=="United States")')
size=np.where(data.Year % 10 == 0,50,10)
color = np.where(data.Country == "Italy",'blue','orange')
data.plot.scatter("gdp","life",logx=True,s=size, c=color)
```

```python
data = gapminder.query('(Country=="China") or (Country=="United States")')
size=np.where(data.Year % 10 == 0,50,10)
color = np.where(data.Country == "China",'red','orange')
data.plot.scatter("gdp","life",logx=True,s=size, c=color)
```

```python
data = gapminder.query('(Country=="China") or (Country=="India")')
size=np.where(data.Year % 10 == 0,50,10)
color = np.where(data.Country == "China",'red','orange')
ax = data.plot.scatter("gdp","life",logx=True,s=size, c=color)
data[data.Country == "China"].plot.line(x='gdp',y='life',ax=ax)
```

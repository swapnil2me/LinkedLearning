# Queue Structure

```python
import queue
import threading
import time
```

```python
serving_line = queue.Queue(maxsize=5)

def soup_producer():
    for i in range(20):
        serving_line.put_nowait('Bowl #'+str(i))
        print('served_Bowl #', str(i), '- remaining capacity:',\
             serving_line.maxsize-serving_line.qsize())
        time.sleep(0.2)
        
def soup_consumer():
    while True:
        bowl = serving_line.get()
        print('Ate', bowl)
        time.sleep(0.3)
```

```python
threading.Thread(target =soup_producer).start()
threading.Thread(target = soup_consumer).start()
```
